```python
"""
Q3. Why is redirect() used in Flask?
Ans.In Flask, the redirect() function is used to perform a redirect in a web application. A redirect is a way to instruct the client's web browser 
    to navigate to a different URL. This can be useful for various purposes in web development, such as after a form submission, after a successful 
    login, or when handling routes that have moved or changed.
    
    Common use cases for the redirect() function in Flask include:
    After Form Submissions: After a user submits a form, you might want to redirect them to a different page to show a confirmation message or 
    display the results.
    
    Changing URLs: If you change the structure of your routes, you can use redirects to ensure that users who visit old URLs are automatically 
    directed to the new ones.
    
    Handling Invalid URLs: When a user tries to access a non-existent or invalid URL, you can use a redirect to guide them to a proper page.
"""
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return redirect(url_for('hello'))

@app.route('/hello')
def hello():
    return 'Hello, World!'

if __name__ == '__main__':
    app.run()
```
