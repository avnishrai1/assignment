```python
"""
Q3. What is App routing in Flask? Why do we use app routes?
Ans.App routing in Flask refers to the process of mapping specific URLs (Uniform Resource Locators) to Python functions within a Flask 
    application. In other words, it's the way Flask determines which function to execute when a user visits a particular URL in the web 
    application.
    
    Flask uses decorators, such as @app.route('/url'), to associate a URL with a Python function (also known as a view function). When a 
    user accesses a URL that matches the route you've defined, Flask calls the associated view function, and the function's response is sent 
    back to the user's browser.
"""

from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/avnish")
def avnish():
    return "Hello World!"

if __name__ == "__main__" :
    app.run(host="0.0.0.0")

"""
In this example, accessing the root URL ("/avnish") will trigger the avnish() function. App routes allow you to control what content is 
displayed on different pages of your web application.
"""
```

```python
"""
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Attach the screenshot of the output in Jupyter Notebook.
"""
from flask import Flask

app = Flask(__name__)

@app.route("/welome")
def welcome() :
    return "Welcome to ABC Corporation"

@app.route("/")
def company():
    data = """
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999 """
    
    return data

if __name__ == "__main__" :
    app.run(host=0.0.0.0)
```

```python
"""
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans.In Flask, the url_for() function is used for URL building. It generates a URL for a given view function, allowing you to avoid 
    hardcoding URLs in your templates or code. This is particularly useful if you later change the route of a view function or if you have
    dynamic URLs with parameters.
"""
from flask import Flask
from flask import url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the home page!"

@app.route('/about')
def about():
    return "This is the about page."

@app.route('/contact')
def contact():
    return "You can contact us here."

if __name__ == '__main__':
    with app.test_request_context():
    
        home_url = url_for('home') #generate for hoem
        about_url = url_for('about') #for about
        contact_url = url_for('contact') #for contact
    
    app.run(host="0.0.0.0")
```
